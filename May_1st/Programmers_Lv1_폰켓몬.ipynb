## 풀이

```python
nums = [3,1,2,3]
n = len(nums)

# 중복되지 않은 리스트 생성 -> set()를 생각하지 못함
num = []
for i in nums:
    if i not in num:
        num.append(i)

# 길이를 비교하여 더 작은 값으로 결과 출력
m = len(num)
if n//2 <= m:
    answer = n//2
else:
    answer = m

answer
```

## 다른 사람의 풀이

```python
answer = min(len(nums)/2, len(set(nums)))
answer
```
