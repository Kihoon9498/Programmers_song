## 풀이

```python
s = 'AB'
n = 1

# 문자열을 미리 정한 후 대문자, 소문자 이용하여 answer 도출
a = 'abcdefghijklmnopqrstuvwxyz'
answer = ''
for word in s:
    if word.isupper():
        answer += a[a.index(word.lower()) + n - len(a)].upper()
    elif word == ' ':
        answer += word
    else:
        answer += a[a.index(word) + n - len(a)]
answer
```

## 다른 사람의 풀이

```python
def solution(s, n):
    s = list(s)
    for i in range(len(s)):
        if s[i].isupper():
            s[i]=chr((ord(s[i])-ord('A')+ n)%26+ord('A'))
        elif s[i].islower():
            s[i]=chr((ord(s[i])-ord('a')+ n)%26+ord('a'))

    return "".join(s)
solution(s, n)
```
