## 풀이

```python
a = 5
b = 24

# 달별 일수를 딕셔너리로 지정해놓고 해당 요일을 계산하려 함
answer = ''
months = {'1':31, '2':29, '3':31, '4':30, '5':31, '6':30, '7':31, '8':31, '9':30, '10':31, '11':30, '12':31}
days = ['SUN', 'MON', 'TUE', 'WED', 'THU', 'FRI', 'SAT']
day = 5
for i in range(1, a):
    day += months[str(i)]
day += b
# 요일을 얻어내는 과정
days[day%7 - 1]
```

## 다른 사람의 풀이

```python
def getDayName(a,b):
    # 리스트로 저장해놓을 생각은 하지 못함
    month = [31, 29, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
    day = ['FRI', 'SAT', 'SUN', 'MON', 'TUE', 'WED', 'THU']
    return day[(sum(month[:a-1])+b-1)%7]

getDayName(5, 24)
```
