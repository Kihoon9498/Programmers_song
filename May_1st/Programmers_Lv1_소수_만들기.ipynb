## 풀이

```python
nums = [1,2,7,6,4]
# 서로 다른 3개의 숫자를 고를 때 인덱스를 이용하였고, 더한 후 소수인지 판별
answer = 0
for i in range(len(nums)):
    for j in range(1, len(nums)-i):
        for k in range(1, len(nums)-i-j):
            num = nums[i] + nums[i+j] + nums[i+j+k]
            answer += 1
            for n in range(2, num):
                if num % n == 0:
                    answer -= 1
                    break
answer
```

## 다른 사람의 풀이

```python
# Combinations 함수를 이용
def solution(nums):
    from itertools import combinations as cb
    answer = 0
    for a in cb(nums, 3):
        cand = sum(a)
        for j in range(2, cand):
            if cand%j==0:
                break
        else:
            answer += 1
    return answer
solution([1,2,7,6,4])
```
