## 풀이

```python
n = 500

# 3~n까지 홀수만 반복하여 해결하고자 하였으나 효율성 문제에서 걸림
answer = 1
for num in range(3, n + 1, 2):
    answer += 1
    for i in range(2, num):
        if num % i == 0:
            answer -= 1
            break
answer
```

## 다른 사람의 풀이

```python
# 에라토스테네스의 체 라는 이론을 활용한 코드
def solution(n):
    answer = set(range(2,n+1))
    
    for i in range(2,n+1):
        if i in answer:
            answer -= set(range(i*2,n+1,i))
    return len(answer)
solution(500)
```
