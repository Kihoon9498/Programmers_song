## 풀이

```python
s = 'banana'
# 문자열을 뒤집어서 반복문을 이용하여 풀고자 함
new_s = s[::-1]
n = len(s)
answer = []
while new_s:
    if new_s[0] in new_s[1:]:
        answer.append(new_s[1:].index(new_s[0])+1)
    else:
        answer.append(-1)
    new_s = new_s[1:]
answer = sorted(answer)
answer
```
