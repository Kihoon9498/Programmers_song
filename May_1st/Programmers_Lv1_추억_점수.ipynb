## 풀이

```python
name = ["may", "kein", "kain", "radi"]
yearning = [5, 10, 1, 3]
photo = [["may", "kein", "kain", "radi"],["may", "kein", "brin", "deny"], ["kon", "kain", "may", "coni"]]
answer = []
# photo 리스트를 반복하며 name의 인덱스 값으로 yearning을 얻어 정답 리스트에 추가하기
for people in photo:
    temp_answer = 0
    for person in people:
        if person in name:
            temp_answer += yearning[name.index(person)]
    answer.append(temp_answer)

answer
```
