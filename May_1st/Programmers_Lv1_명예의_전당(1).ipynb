## 풀이

```python
score = [0, 300, 40, 300, 20, 70, 150, 50, 500, 1000]
k = 4

answer = []
temp_list = []

for i in score:
    if len(temp_list) < k:
        temp_list.append(i)
        temp_list = sorted(temp_list)
    else:
        if temp_list[0] <= i:
            temp_list[0] = i
            temp_list = sorted(temp_list)
    answer.append(temp_list[0])
answer
```
