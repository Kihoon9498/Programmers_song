## 풀이

```python
answers = [1,3,2,4,2]

# 패턴을 담은 리스트 생성
lis1 = [1,2,3,4,5]
lis2 = [2,1,2,3,2,4,2,5]
lis3 = [3,3,1,1,2,2,4,4,5,5]
score_list = [0, 0, 0]
answer = []
# 인덱스, 값을 이용하여 정답과 맞추면 스코어를 올림
for n, i in enumerate(answers):
    if i == lis1[n % len(lis1)]:
        score_list[0] += 1
    if i == lis2[n % len(lis2)]:
        score_list[1] += 1
    if i == lis3[n % len(lis3)]:
        score_list[2] += 1
# 점수가 제일 높은 사람
max_score = max(score_list)
if score_list.count(max_score) == 1:
    answer.append(score_list.index(max_score)+1)
else:
    for n, i in enumerate(score_list):
        if i == max_score:
            answer.append(n+1)
answer
```

## 다른 사람의 풀이

```python
def solution(answers):
    p = [[1, 2, 3, 4, 5],
         [2, 1, 2, 3, 2, 4, 2, 5],
         [3, 3, 1, 1, 2, 2, 4, 4, 5, 5]]
    s = [0] * len(p)

    for q, a in enumerate(answers):
        for i, v in enumerate(p):
            if a == v[q % len(v)]:
                s[i] += 1
    return [i + 1 for i, v in enumerate(s) if v == max(s)]

solution([1,3,2,4,2])
```
