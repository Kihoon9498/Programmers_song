## 풀이

```python
N = 5
stages = [2, 1, 2, 6, 2, 4, 3, 3]
m = len(stages)
stages.sort()
stage_list, stage_ratio, answer = [], [], []
# 길이가 N+1인 0행렬을 만듦
for i in range(N+1):
    stage_list.append(0)
# stages에서 stage_list의 인덱스에 맞는 값을 더함
for i in stages:
    stage_list[i-1] += 1
# stages의 길이로 나누어서 비율을 계산함
for i in stage_list[:-1]:
    try:
        stage_ratio.append(i / m)
        m -= i
    except:
        stage_ratio.append(0)
# 비율의 인덱스를 이용하여 순위를 계산함
while len(stage_ratio) != stage_ratio.count(-1):
    answer.append(stage_ratio.index(max(stage_ratio))+1)
    stage_ratio[stage_ratio.index(max(stage_ratio))] = -1

answer
```

## 다른 사람의 풀이

```python
def solution(N, stages):
    result = {}
    denominator = len(stages)
    for stage in range(1, N+1):
        if denominator != 0:
            count = stages.count(stage)
            result[stage] = count / denominator
            denominator -= count
        else:
            result[stage] = 0
    return sorted(result, key=lambda x : result[x], reverse=True)
solution(N, stages)
```
