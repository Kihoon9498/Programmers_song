## 풀이

```python
lottos = [44, 1, 0, 0, 31, 25]
win_nums = [31, 10, 45, 1, 6, 19]
# 맞춘 개수와 등수를 딕셔너리 형태로 저장
win_dict = {0:6, 1:6, 2:5, 3:4, 4:3, 5:2, 6:1}
n = lottos.count(0)
counts = 0
for i in lottos:
    if i in win_nums:
        counts += 1
# 최저, 최고 순위 얻기
answer = [win_dict[counts+n], win_dict[counts]]
answer
```

## 다른 사람의 풀이

```python
def solution(lottos, win_nums):

    rank=[6,6,5,4,3,2,1]

    cnt_0 = lottos.count(0)
    ans = 0
    for x in win_nums:
        if x in lottos:
            ans += 1
    return rank[cnt_0 + ans],rank[ans]

solution(lottos, win_nums)
```
