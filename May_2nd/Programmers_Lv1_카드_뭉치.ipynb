## 풀이

```python
cards1 = ["i", "water", "drink"]
cards2 = ["want", "to"]
goal = ["i", "want", "to", "drink", "water"]

# 리스트의 첫번째 원소들을 확인하여 동일하면 순서대로 제거
while goal:
    if (goal[0] in cards1) and (goal[0] == cards1[0]):
        goal.remove(goal[0])
        cards1.remove(cards1[0])
    elif (goal[0] in cards2) and (goal[0] == cards2[0]):
        goal.remove(goal[0])
        cards2.remove(cards2[0])
    else:
        break
answer = 'Yes' if bool(not goal) else 'No'
answer
```

## 다른 사람의 풀이

```python
def solution(cards1, cards2, goal):
    for g in goal:
        if len(cards1) > 0 and g == cards1[0]:
            cards1.pop(0)       
        elif len(cards2) >0 and g == cards2[0]:
            cards2.pop(0)
        else:
            return "No"
    return "Yes"

solution(cards1, cards2, goal)
```
